```python
import re

# Define the regex pattern and replacement function
pattern = r"(={1,6})\s*([^=]+?)\s*\1"
replacement = lambda match: f"\n\n{match.group(0)}\n\n"

# Example 1: Equal signs exactly next to the title (inline)
example_1 = """
Lorem ipsum dolor sit amet, ===Ferguson years (1986–2013)=== consectetur adipiscing elit.
"""

# Example 2: Spaces around the title (inline)
example_2 = """
Curabitur blandit tempus porttitor. === Ferguson years (1986–2013) === Vivamus sagittis lacus vel augue.
"""

# Example 3: Space only at the end of the title (inline)
example_3 = """
Nullam quis risus eget urna mollis ornare vel eu leo. =Title with space at end = Maecenas sed diam eget risus.
"""

# Example 4: Invalid header (different number of equal signs, inline)
example_4 = """
Donec id elit non mi porta gravida at eget metus. === Invalid Header == Cras justo odio, dapibus ac facilisis in.
"""

# Example 5: Equal signs exactly next to the title (short, inline)
example_5 = """
Aenean lacinia bibendum nulla sed consectetur. ==Valid== Etiam porta sem malesuada magna mollis euismod.
"""

# Example 6: Spaces around the title (short, inline)
example_6 = """
Praesent commodo cursus magna, vel scelerisque nisl. = Also Valid = Nullam id dolor id nibh ultricies vehicula ut id elit.
"""

# Example 7: No title text, just equal signs (inline)
example_7 = """
Fusce dapibus, tellus ac cursus commodo. ======= Tortor mauris condimentum nibh, ut fermentum massa.
"""

# Example 8: Mixed case and special characters (inline)
example_8 = """
Vestibulum id ligula porta felis euismod semper. === fERGUSON yEARS (1986–2013) & Beyond! === Cum sociis natoque penatibus.
"""

# Example 9: Header within a paragraph (inline)
example_9 = """
This is a paragraph with a header === Ferguson years (1986–2013) === inside it, and some more text.
"""

# Example 10: Multiple headers (inline)
example_10 = """
Lorem ipsum dolor sit amet, === Ferguson years (1986–2013) === consectetur adipiscing elit. == Another Title == Pellentesque habitant morbi.
"""

# Test all examples
examples = [example_1, example_2, example_3, example_4, example_5, example_6, example_7, example_8, example_9, example_10]

for i, example in enumerate(examples, 1):
    print(f"Example {i} Before:")
    print(example)
    final_text = re.sub(pattern, replacement, example)
    print(f"Example {i} After:")
    print(final_text)
    print("-" * 40)
```

```python
import re

# Your input text
text = """= Manchester United F.C. =
Manchester United Football Club (F.C.) is a professional football club. It is based in Old Trafford, Greater Manchester, England. It plays in the Premier League, the highest level of English football. Its nickname is "the Red Devils". The club started as Newton Heath LYR Football Club, in 1878. It changed its name to Manchester United in 1902. It moved to its current stadium, Old Trafford, in 1910.
== History ==
=== Early years (1878–1945) ===
Manchester United was started in 1878. It was called Newton Heath LYR Football Club.
= Related pages =
* List of Manchester United F.C. players
= Notes =
= References ="""

# Regex pattern to match valid titles enclosed in = signs
pattern = r"(^|\n)(={1,6})\s*([^=\n]+?)\s*\2(\n|$)"

# Replacement function to add newlines before and after the title
replacement = lambda match: f"{match.group(1)}\n\n{match.group(2)}{match.group(3)}{match.group(2)}\n\n{match.group(4)}"

# Apply the regex substitution
result = re.sub(pattern, replacement, text)

print(result)
```
